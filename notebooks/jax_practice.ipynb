```python
import jax.numpy as jnp
import jax
```

```python
key = jax.random.PRNGKey(0)
keys = jax.random.split(key, 4)

xdim = 5
ydim = 10

# Latent params
Q = jnp.eye(xdim)
Q0 = jnp.eye(xdim)
A = jnp.eye(xdim)
T = 100

# Observation params
R = jnp.eye(ydim)
C = jax.random.normal(keys[1], (ydim, xdim))

def cov_step(Pt, _):
    P_pred = A @ Pt @ A.T + Q   # need to handle t==0 case (see below)
    P_obs = C @ P_pred @ C.T + R
    P_obs_inv = jnp.linalg.inv(P_obs)
    _, log_det = jnp.linalg.slogdet(P_obs)
    K = P_pred @ C.T @ P_obs_inv
    Pt_new = P_pred - K @ C @ P_pred
    return Pt_new, (P_pred, P_obs_inv, log_det, K, Pt_new)

_, (P_pred_all, P_obs_inv_all, log_det_all, K_all, P_filt_all) = jax.lax.scan(
    cov_step, init=Q0, xs=None, length=T,
)
```
